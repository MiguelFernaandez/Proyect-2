```python
import pandas as pd
import numpy as ny
import pymysql
import sqlalchemy as alch 
from getpass import getpass
import re
```

```python
credit_card_db =  pd.read_csv("CC_trans.csv")
credit_card_db
```

```python
credit_card_db.drop(columns='index', inplace=True)
```

```python
credit_card_db
```

```python
credit_card_db.insert(0, 'trans_id', range(0, len(credit_card_db)))
credit_card_db
```

```python
card_type_db = credit_card_db[['trans_id', 'Card Type']]
card_type_db
```

```python
trans_db = credit_card_db[['trans_id', 'Card Type', "Exp Type", "Amount", "Gender", "City", "Date"]]
trans_db
```

```python
credit_card_db["Card Type"].value_counts()
```

```python


patrones_card = {
    'Silver': 1,
    'Signature': 2,
    'Platinum': 3,
    'Gold': 4
}

for pattern, value in patrones_card.items():
    trans_db['Card Type'] = trans_db['Card Type'].str.replace(pattern, str(value))
```

```python
trans_db
```

```python
trans_db['Exp Type'].value_counts()
```

```python
card_type_db['Card_id'] = card_type_db['Card Type'].map(patrones_card)
```

```python
card_type_db.drop(columns='trans_id', inplace=True)
card_type_db
```

```python
patrones_exp = {
    'Food': 1,
    'Fuel': 2,
    'Bills': 3,
    'Entertainment': 4,
    'Grocery': 5,
    'Travel': 6
}

trans_db['Exp_id'] = trans_db['Exp Type'].map(patrones_exp)

trans_db.drop(columns='Exp Type', inplace=True)
```

```python
patrones_exp = {
    'Food': 1,
    'Fuel': 2,
    'Bills': 3,
    'Entertainment': 4,
    'Grocery': 5,
    'Travel': 6
}

credit_card_db['Exp_id'] = credit_card_db['Exp Type'].map(patrones_exp)

```

```python
trans_db
```

```python
expense_db = credit_card_db[["Exp_id", "Exp Type"]]

expense_db = expense_db.drop_duplicates()
expense_db
```

```python
card_type_db = card_type_db.drop_duplicates()
card_type_db
```

```python
def cambio_columnas(df):
    return df.rename(columns=lambda x: x.replace(' ', '_'))

card_type_db = cambio_columnas(card_type_db)
expense_db = cambio_columnas(expense_db)

print(f'{trans_db.keys()} \n {card_type_db.keys()}\n {expense_db.keys()}')
```

```python
trans_db = cambio_columnas(trans_db)
trans_db
```

```python
trans_db.to_csv("trans_table.csv", index=False)
expense_db.to_csv("expense_table.csv", index=False)
card_type_db.to_csv("card_type_table.csv", index=False)
```
